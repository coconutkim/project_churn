```python
bins=[300,600,660,720,780,850]
labels=['Very poor','Poor','Fair','Good','Excellent']
```

```python
bank_df['section of credit_score'] = pd.cut(
    bank_df['credit_score'], bins=bins, labels=labels,right=True)
```

```python
bank_df
```

```python
bank_df.insert(bank_df.columns.get_loc('credit_score') + 1,
              'section of credit_score',
              bank_df.pop('section of credit_score'))
```
