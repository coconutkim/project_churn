```python
bins=[300,600,660,720,780,850]
labels=['Very poor','Poor','Fair','Good','Excellent']
```

```python
bank_df['section of credit_score'] = pd.cut(
    bank_df['credit_score'], bins=bins, labels=labels,right=True)
```

```python
bank_df
```

```python
bank_df.insert(bank_df.columns.get_loc('credit_score') + 1,
              'section of credit_score',
              bank_df.pop('section of credit_score'))
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 'df'는 이미 정의되고 로드되었다고 가정합니다

# 'churn'이 1인 데이터만 필터링합니다
filtered_df = bank_df[bank_df['churn'] == 1]

# 'country' 열의 값 분포를 집계합니다
distribution = filtered_df['section of credit_score'].value_counts()

# 막대 그래프를 생성합니다
plt.figure(figsize=(8, 6))
ax = distribution.plot(kind='bar', color='skyblue', edgecolor='black')

# 막대 위에 퍼센트를 표시합니다
total = distribution.sum()  # 전체 항목 수를 계산합니다
for p in ax.patches:  # 각 막대를 순회합니다
    height = p.get_height()  # 막대의 높이(빈도수)를 가져옵니다
    percentage = (height / total) * 100  # 퍼센트를 계산합니다
    ax.annotate(f'{percentage:.1f}%',  # 퍼센트 값을 막대 위에 표시합니다
                (p.get_x() + p.get_width() / 2., height),  # 텍스트의 위치를 설정합니다
                ha='center', va='center',  # 텍스트 정렬 설정
                xytext=(0, 3),  # 텍스트 위치 조정
                textcoords='offset points')

# 제목과 축 레이블을 설정합니다
# plt.title('Distribution of Countries with Churn = 1')  # 제목을 영어로 설정합니다
plt.xlabel('section of credit_score')  # X축 레이블 설정
plt.ylabel('Frequency')  # Y축 레이블 설정
plt.xticks(rotation=45)  # X축 레이블을 45도 회전시킵니다
plt.tight_layout()  # 레이아웃을 자동으로 조정합니다

# 그래프를 표시합니다
plt.show()
```
