```python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency

# Cramér's V 계산 함수
def cramers_v(x, y):
    confusion_matrix = pd.crosstab(x, y)
    chi2 = chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    r, k = confusion_matrix.shape
    return np.sqrt(chi2 / (n * (min(r, k) - 1)))

# 5개의 범주형 칼럼 선택 (칼럼명에 맞게 변경)
selected_columns = ['country', 'gender', 'credit_card', 'active_member', 'churn']

# 빈 매트릭스 생성
cramers_v_matrix = pd.DataFrame(np.zeros((len(selected_columns), len(selected_columns))), 
                                index=selected_columns, columns=selected_columns)

# Cramér's V 계산 및 매트릭스 채우기
for i in range(len(selected_columns)):
    for j in range(i + 1, len(selected_columns)):
        cramers_v_matrix.iloc[i, j] = cramers_v(df_exceptID[selected_columns[i]], df_exceptID[selected_columns[j]])
        cramers_v_matrix.iloc[j, i] = cramers_v_matrix.iloc[i, j]

# 결과 출력
print(cramers_v_matrix)

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 히트맵 생성
plt.figure(figsize=(10, 8))
sns.heatmap(cramers_v_matrix, annot=True, cmap='YlGnBu', cbar=True, linewidths=0.5, linecolor='gray')

# 제목 및 레이블 추가
plt.title('Heatmap of Numeric Data')
plt.xlabel('Columns')
plt.ylabel('Index')

# 히트맵 표시
plt.show()

```

```python
import pandas as pd

# 예시: Cramér's V 계산 결과가 담긴 데이터프레임
# cramers_v_df = pd.DataFrame({
#     'Category1': [0, 0.2, 0.3, 0.4],
#     'Category2': [0.2, 0, 0.1, 0.2],
#     'Category3': [0.3, 0.1, 0, 0.2],
#     'Category4': [0.4, 0.2, 0.2, 0]
# }, index=['Category1', 'Category2', 'Category3', 'Category4'])

# Cramér's V가 0.1 이상인 변수 쌍 추출
threshold = 0.1
significant_pairs = []

# 데이터프레임의 상삼각 행렬 부분을 대상으로 상관관계가 있는 변수 쌍 추출
for i in range(len(cramers_v_matrix.columns)):
    for j in range(i + 1, len(cramers_v_matrix.columns)):
        var1 = cramers_v_matrix.columns[i]
        var2 = cramers_v_matrix.columns[j]
        value = cramers_v_matrix.iloc[i, j]
        if value >= threshold:
            significant_pairs.append((var1, var2, value))

print("Significant variable pairs (Cramér's V >= 0.1):")
for pair in significant_pairs:
    print(f"Variables: {pair[0]} and {pair[1]}, Cramér's V: {pair[2]}")

```

```python
%matplotlib inline
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# 수치형 변수만 추출
numeric_df = df_exceptID.select_dtypes(include=['number'])

# 상관계수 행렬 계산
corr_matrix = numeric_df.corr()

# 상삼각형 마스크 생성
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

# 히트맵 시각화
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, mask=mask, annot=True, cmap='coolwarm',
            fmt='.2f', linewidths=0.5, square=True)
plt.title('Correlation Heatmap (Upper Triangle)')
plt.show()

# 오류사항
# 상관계수는 수치형 데이터를 이용해야 한다
# 범주형 데이터는 해당되지 않는다
```
