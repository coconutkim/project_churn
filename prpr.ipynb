```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 박스플롯 시각화
plt.figure(figsize=(10, 6))
sns.boxplot(data=df)
plt.title('Boxplot of Features')
plt.show()
```

```python
# 히스토그램과 KDE 시각화
df.plot(kind='kde', subplots=True, figsize=(10, 8), sharex=False)
plt.suptitle('KDE Plots of Features')
plt.show()
```

```python
# describe() 요약 통계량을 바 차트로 시각화
desc.T.plot(kind='bar', figsize=(10, 6))
plt.title('Summary Statistics')
plt.ylabel('Value')
plt.xticks(rotation=0)
plt.show()
```
