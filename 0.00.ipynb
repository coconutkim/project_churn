```python
df['balance_group'] = df['balance_group'].astype('category')
```

```python
df['balance_group'].dtype
```

```python
# 'A' 칼럼의 값이 'zero'인 행만 필터링
df_filtered = df[df['balance_group'] == 'zero']
```

```python
df_filtered
```

```python
# 'B' 칼럼 제거
df_dropped = df_filtered.drop(columns=['customer_id'])
df_dropped
```

```python
df_dropped.dtypes
```

```python
df_dropped['section of credit_score'] = df_dropped['section of credit_score'].astype('category')
df_dropped['country'] = df_dropped['country'].astype('category')
df_dropped['gender'] = df_dropped['gender'].astype('category')
df_dropped['age_group'] = df_dropped['age_group'].astype('category')
df_dropped['active_member'] = df_dropped['active_member'].astype('category')
df_dropped['churn'] = df_dropped['churn'].astype('category')
df_dropped['credit_card'] = df_dropped['credit_card'].astype('category')
```

```python
df_dropped.dtypes
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 'category' 타입 칼럼만 선택
category_columns = df_dropped.select_dtypes(include=['category'])

# 각 'category' 타입 칼럼에 대해 막대 그래프 그리기
for col in category_columns.columns:
    # 값의 빈도수를 계산
    freq = category_columns[col].value_counts()
    
    # 막대 그래프 그리기
    plt.figure(figsize=(8, 6))
    freq.plot(kind='bar', color='skyblue')
    plt.title(f'{col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.xticks(rotation=45)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()
```
