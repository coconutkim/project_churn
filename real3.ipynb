```python
import pandas as pd

df=pd.read_csv('./grouped_data3.csv')
df
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.yticks([0, 1])

plt.scatter(df['products_number'],df['balance'])
plt.xlabel('products_number')
plt.xticks([1,2,3,4])
plt.ylabel('balance')
plt.yticks(np.arange(0, 250001, 50000))
plt.title('Scatter Plot')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# 이탈 여부 변수의 카운트 계산
churn_counts = df['churn'].value_counts()

# 파이 그래프 그리기
plt.figure(figsize=(8, 6))
plt.pie(churn_counts, labels=churn_counts.index,
        autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff'])

# 제목 설정
plt.title('Customer Churn Distribution')

# 그래프 표시
plt.show()
```
