```python
import pandas as pd

df=pd.read_csv('./grouped_data3.csv')
df
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.yticks([0, 1])

plt.scatter(df['products_number'],df['balance'])
plt.xlabel('products_number')
plt.xticks([1,2,3,4])
plt.ylabel('balance')
plt.yticks(np.arange(0, 250001, 50000))
plt.title('Scatter Plot')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# 이탈 여부 변수의 카운트 계산
churn_counts = df['churn'].value_counts()

# 파이 그래프 그리기
plt.figure(figsize=(8, 6))
plt.pie(churn_counts, labels=churn_counts.index,
        autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff'])

# 제목 설정
plt.title('Customer Churn Distribution')

# 그래프 표시
plt.show()
```

```python
# 추정 봉급 변수에서 이탈자와 무이탈자의 상관관계
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='churn', y='age')
plt.title('Box Plot of Estimated Salary by Churn Status')
plt.xlabel('churn')
plt.ylabel('age')
plt.xticks([0, 1], ['0', '1'])
plt.show()
```

```python
# 연봉 구간 정의
# 낮은 연봉: 25,000 - 35,000 유로
# 중간 연봉: 35,000 - 60,000 유로
# 높은 연봉: 60,000 - 100,000 유로
# 매우 높은 연봉: 100,000 유로 이상
```

```python
print(round((df['estimated_salary'].min()*12),3))
print(round((df['estimated_salary'].max()*12),2))
```

```python
# 추정 봉급의 섹션을 사분위수가 아니라 일반적인 재산 나누는 기준으로 해본다

bins = [0,35000,60000,100000,float('inf')]
labels = ['low', 'medium', 'high', 'very high']
df['salary_range'] = pd.cut(df['estimated_salary'], bins=bins,
                            labels=labels, include_lowest=True)
df
```

```python
# 일단 데이터 프레임 정리를 좀 한다
# 복잡한 변수 이름을 데이터 값을 예상하기 쉽고, 단순한 것으로 바꾼다

# 변수 이름 바꾸기
# df = df.rename(columns={'old_name': 'new_name'})

df = df.rename(columns={'section of credit_score': 'credit__section',
                        'agegroup_gender': 'age_group_gender',
                        'age_balance': 'balance_per_age'})
```
