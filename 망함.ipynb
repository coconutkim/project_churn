```python
import pandas as pd
import matplotlib.pyplot as plt

# 'df'는 이미 정의되고 로드되었다고 가정합니다

# 'churn'이 1인 데이터만 필터링합니다
filtered_df = bank_df[bank_df['churn'] == 1]

# 'country' 열의 값 분포를 집계합니다
distribution = filtered_df['section of credit_score'].value_counts()

# 막대 그래프를 생성합니다
plt.figure(figsize=(8, 6))
ax = distribution.plot(kind='bar', color='skyblue', edgecolor='black')

# 막대 위에 퍼센트를 표시합니다
total = distribution.sum()  # 전체 항목 수를 계산합니다
for p in ax.patches:  # 각 막대를 순회합니다
    height = p.get_height()  # 막대의 높이(빈도수)를 가져옵니다
    percentage = (height / total) * 100  # 퍼센트를 계산합니다
    ax.annotate(f'{percentage:.1f}%',  # 퍼센트 값을 막대 위에 표시합니다
                (p.get_x() + p.get_width() / 2., height),  # 텍스트의 위치를 설정합니다
                ha='center', va='center',  # 텍스트 정렬 설정
                xytext=(0, 3),  # 텍스트 위치 조정
                textcoords='offset points')

# 제목과 축 레이블을 설정합니다
# plt.title('Distribution of Countries with Churn = 1')  # 제목을 영어로 설정합니다
plt.xlabel('section of credit_score')  # X축 레이블 설정
plt.ylabel('Frequency')  # Y축 레이블 설정
plt.xticks(rotation=45)  # X축 레이블을 45도 회전시킵니다
plt.tight_layout()  # 레이아웃을 자동으로 조정합니다

# 그래프를 표시합니다
plt.show()
```

```python
# 우선 데이터의 상한선을 찾아 어디서부터 이상치로 판별할 수 있는지 알아본다

import numpy as np

# 데이터 정렬
data_sorted = np.sort(filtered_df['age'])

# 1사분위수(Q1)와 3사분위수(Q3) 계산
Q1 = np.percentile(data_sorted, 25)
Q3 = np.percentile(data_sorted, 75)

# IQR 계산
IQR = Q3 - Q1

# 상한선 계산
upper_bound = Q3 + 1.5 * IQR

print(f"1사분위수 (Q1): {Q1}")
print(f"3사분위수 (Q3): {Q3}")
print(f"IQR: {IQR}")
print(f"상한선: {upper_bound}")

# 상한선 초과 데이터 확인
outliers_above = [x for x in filtered_df['age'] if x > upper_bound]
print(f"상한선 초과 데이터: {outliers_above}")

# 상한선: 70.5
# 상한선 초과 데이터: [73, 71, 74, 71, 84, 71, 71, 71, 71, 72, 71]

# 상한선을 넘긴 하지만 범위를 지나치게 벗어나지 않음
# 이상치로 간주하지 않는다

```

```python
# 각각의 나이대에서 1과 0의 비율 계산하기
# 예) 10대 1:~명 0:~명
나이 변수에서 

# 나이대별로 ㄴ의 값이 0 또는 1인 경우의 개수 계산
grouped = filtered_df.groupby('age_group')['churn'].value_counts().unstack().fillna(0)

# 비율 계산
grouped['total'] = grouped[0] + grouped[1]  # 전체 개수
grouped['0_ratio'] = grouped[0] / grouped['total']  # 0의 비율
grouped['1_ratio'] = grouped[1] / grouped['total']  # 1의 비율

print(grouped[['0_ratio', '1_ratio']])
```
